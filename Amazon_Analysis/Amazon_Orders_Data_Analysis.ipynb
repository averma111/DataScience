```python
import pandas as pd
import numpy as np
```

```python
# Read amazon data in csv format

amazon_df = pd.read_csv('data/amazon_data.csv')
amazon_df
```

```python
amazon_df.shape
```

```python
# Cleaning the NAN values
amazon_df.fillna(0)
```

```python
amazon_df.columns
```

```python
amazon_df['Total Charged'] = amazon_df['Total Charged'].str.replace('$','').astype(float)
```

```python
amazon_df['Total Charged']
```

```python
# Finding the total amount spent on amazon orders
amazon_df['Total Charged'].sum()

#11207.64
```

```python
#Finding the meam value of amount spent
amazon_df['Total Charged'].mean()
```

```python
#Finding the meam value of amount spent
amazon_df['Total Charged'].median()
```

```python
amazon_df['Tax Charged'] = amazon_df['Tax Charged'].str.replace('$','').astype(float)
```

```python
# Total tax charged from buying goods from amazon
amazon_df['Tax Charged'].sum()
```

```python
amazon_df['Order Date'] = pd.to_datetime(amazon_df['Order Date'])
```

```python
%matplotlib inline
```

```python
yearly_orders = amazon_df.groupby(amazon_df['Order Date'].map(lambda x: x.year)).sum()["Total Charged"]
```

```python
yearly_orders
```

```python
yearly_orders.plot.bar(figsize=(20,10))
```

```python
monthly_orders = amazon_df.groupby(amazon_df['Order Date'].map(lambda x: x.month)).sum()["Total Charged"]
```

```python
monthly_orders.plot.bar(figsize=(20,10))
```
